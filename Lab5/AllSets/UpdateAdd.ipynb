```python
import math
from tkinter import *
import re
import sqlite3
root = Tk()
root.geometry('850x850')
root.title("STUDENT REGISTRATION FORM")
def database():
    first_name=entry1.get()
    last_name=entry2.get()
    email=entry3.get()
    
    conn=sqlite3.connect('form1.db')
    with conn:
                cursor=conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS  Student_reg (first_name TEXT,last_name TEXT,email TEXT)')
    cursor.execute('INSERT INTO Student_reg (first_name,last_name,email) VALUES(?,?,?)',(first_name,last_name,email))
    validation()
    conn.commit()







def check(str):
    if (str.isalpha() and length(str)<=30) or (str.isalpha() and length(str)<=10):
        return str
    elif str=="":
        return "FIELD MISSING"
    else:
        return "INVALID INPUT"
    
def update():
    first_name=entry1.get()
    last_name=entry2.get()
    email=entry3.get()
    conn=sqlite3.connect("form1.db")
    with conn:
        cursor=conn.cursor()
    cursor.execute('UPDATE Student_reg SET first_name=?, last_name=?, email=?',(first_name,last_name,email))
    conn.commit()
    conn.close()
    conn=sqlite3.connect("form1.db")
    cursor=conn.cursor()
    cursor.execute("SELECT * FROM Student_reg")
    ans = cursor.fetchall()
    
    print(ans)
   
def validation():
    first_name=entry1.get()
    print("FIRST NAME:",check(first_name))
    last_name=entry2.get()
    print("LAST NAME:",check(last_name))
    email=entry3.get()
    print("EMAIL ID:",check(email))
    
    
    
    
    

def delete():
    entry1.delete(0,'end')
    entry2.delete(0,'end')
    entry3.delete(0,'end')
   
    
label_0 = Label(root, text="STUDENT REGISTRATION FORM",foreground="blue",background="white",width=100,font=("bold", 20))
label_0.place(x=0,y=0)
root.configure(background = "blue")
label1=Label(root,text="FIRST NAME",foreground="white",background="blue",font=("bold",10))
label1.place(x=0,y=70)
entry1=Entry(root,width=30)
entry1.place(x=150,y=70)
label2=Label(root,text="LAST NAME",foreground="white",background="blue",font=("bold",10))
label2.place(x=0,y=100)
entry2=Entry(root,width=30)
entry2.place(x=150,y=100)
label3=Label(root,text="Registration number",foreground="white",background="blue",font=("bold",10))
label3.place(x=0,y=130)
entry3=Entry(root,width=30)
entry3.place(x=150,y=130)
################################################


MyButton1 = Button(root, text="Submit", width=10,fg="black",bg="white",font=("bold",10),bd=2,command=database)
MyButton1.place(x=350,y=755)
MyButton2 = Button(root, text="Reset", width=10,fg="black",bg="white",font=("bold",10),bd=2,command=delete)
MyButton2.place(x=450,y=755)
MyButton3 = Button(root, text="Update", width=10,fg="black",bg="white",font=("bold",10),bd=2,command=update)                
MyButton3.place(x=550,y=755)                


root.mainloop()
import sqlite3
my_conn = sqlite3.connect('form1.db')
###### end of connection ####

##### tkinter window ######
import tkinter  as tk 
from tkinter import * 
my_w = tk.Tk()
my_w.geometry("1000x1000") 

r_set=my_conn.execute('''SELECT * from Student_reg ''');
i=0 # row value inside the loop 
for student in r_set: 
    for j in range(len(student)):
        e = Entry(my_w, width=10, fg='blue') 
        e.grid(row=i, column=j) 
        e.insert(END, student[j])
    i=i+1
my_w.mainloop()
```
