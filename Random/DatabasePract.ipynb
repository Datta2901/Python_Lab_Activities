```python
from tkinter import *
import tkinter as tk
import sqlite3

def printing():
    file1 = open("MyFile.txt", "a") 
    file1.write("First Name " + str(fname.get()))
    file1.write("\n")
    file1.write("Last Name " + str(lname.get()))
    file1.write("\n")
    file1.write("email " + str(mail.get()))
    file1.write("\n")
    file1.write("City " + str(city.get()))
    file1.write("\n")

    file1.write("Phone number " + str(phone.get()))
    file1.write("\n")

    file1.write("State " + str(state.get()))
    file1.write("\n")
    file1.write("Pincode " + str(pincode.get()))
    file1.close()

def submit():
    printing()# printing in a file
    fname.delete(0,END)
    lname.delete(0,END)
    mail.delete(0,END)
    city.delete(0,END)
    phone.delete(0,END)
    state.delete(0,END)
    pincode.delete(0,END)

def Display():
    myconn = sqlite3.Connection("address.db")
    c = myconn.cursor()
    c.execute()
    

    myconn.commit()
    myconn.close()




window = Tk()
window.geometry('400x400')
window.configure(bg = 'red')
mycon = sqlite3.connection("address.db")

f_name = Label(window,text = "First Name : ",fg = 'blue',bg = 'red').grid(row = 0,column = 1,padx = 5,pady = 5)
l_name = Label(window,text = "Last Name : ",fg = 'blue',bg = 'red').grid(row = 1,column = 1,padx = 5,pady = 5)
ma_il = Label(window,text = "Email : ",fg = 'blue',bg = 'red').grid(row = 2,column = 1,padx = 5,pady = 5)
ci_ty = Label(window,text = "City : ",fg = 'blue',bg = 'red').grid(row = 4,column = 1,padx = 5,pady = 5)
ph_one = Label(window,text = "Phone number : ",fg = 'blue',bg = 'red').grid(row = 5,column = 1,padx = 5,pady = 5)
st_ate = Label(window,text = "State : ",fg = 'blue',bg = 'red').grid(row = 6,column = 1,padx = 5,pady = 5)
pin_code = Label(window,text = "Pincode : ",fg = 'blue',bg = 'red').grid(row = 7,column = 1,padx = 5,pady = 5)

# if we use place function we can write in one line
# if we use gid we should not write in a single line 
# as it results in a attribut error if we try to access the value
fname = Entry(window,width=30)
fname.grid(row = 0,column= 2, padx= 5,pady = 5)
lname = Entry(window,width=30)
lname.grid(row = 1,column= 2, padx= 5,pady = 5)
mail = Entry(window,width=30)
mail.grid(row = 2,column= 2, padx= 5,pady = 5)
city = Entry(window,width=30)
city.grid(row = 4,column= 2, padx= 5,pady = 5)
phone = Entry(window,width=30)
phone.grid(row = 5,column= 2, padx= 5,pady = 5)
state = Entry(window,width=30)
state.grid(row = 6,column= 2, padx= 5,pady = 5)
pincode = Entry(window,width=30)
pincode.grid(row = 7,column= 2, padx= 5,pady = 5)



SubmitBtn = Button(text =  "Submit",bg = "blue",fg = "black",command = submit).grid(row = 11,columnspan = 10,rowspan = 5)
DisplayBtn = Button(text =  "Display",bg = "blue",fg = "black",command = Display).grid(row = 11,column = 7,rowspan = 2)



window.mainloop()
```
